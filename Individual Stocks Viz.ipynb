## One-day Returns

```python
# Viz 1
import matplotlib.pyplot as plt

ret_1 = pd.read_csv('Full_Model_1_Raw.csv', index_col = 0)
data_1 = []
for i in range(10):
    data_1.append(json.loads(ret_1.loc[i, 'GARCH_alpha']))
    
fig = plt.figure(figsize =(8, 7)) 
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot(data_1) 
plt.title('GARCH alpha for one-day return')
plt.xlabel('Decile')
plt.ylabel('Alpha coefficient')
plt.show()
```
