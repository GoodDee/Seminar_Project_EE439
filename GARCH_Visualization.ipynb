```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
GJR_3 = [0.3170, 0.2236, 0.5712]
GJR_5 = [0.5708, 0.1649, 0.3467]
GJR_10 = [0.7797, 0.0183, 0.1877]
GJR_14 = [0.8429, 0.0558, 0.1097]
GJR_20 = [0.8622, 0.0276, 0.1120]
```

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, [GJR_3[0], GJR_5[0], GJR_10[0], GJR_14[0], GJR_20[0]], WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('GJR-GARCH Coefficients')
plt.title('GJR-GARCH: alpha')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, [GJR_3[1], GJR_5[1], GJR_10[1], GJR_14[1], GJR_20[1]], WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('GJR-GARCH Coefficients')
plt.title('GJR-GARCH: gamma')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, [GJR_3[-1], GJR_5[-1], GJR_10[-1], GJR_14[-1], GJR_20[-1]], WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('GJR-GARCH Coefficients')
plt.title('GJR-GARCH: beta')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```
