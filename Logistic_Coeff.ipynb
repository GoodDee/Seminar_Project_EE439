```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MACD_neg = [-0.0025, -0.0035, -0.0007, 0.0004, -0.0006]
MOM_neg = [-0.7593, 0.3091, -1.1170, -0.7598, -0.4481]
STD_neg = [7.96e-05, -0.0006, 0.0001, 0.0011, 0.0010]
vol_neg = [0.0025, -0.0051, 0.0117, -0.0007, -0.0149]

MACD_zero = [0.0020, 0.0016, 0.0010, 0.0006, 0.0014]
MOM_zero = [0.9133, 0.2325, -0.1182, -0.2365, -0.1299]
STD_zero = [-0.0045, -0.0017, 0.0007, -0.0021, -0.0016]
vol_zero = [0.0253, 0.0112, -0.0268, -0.0212, -0.0146]

MACD_pos = [0.0005, 0.0019, -0.0003, -0.0009, -0.0008]
MOM_pos = [-0.1539, -0.5415, 1.2352, 0.9963, 0.5780]
STD_pos = [0.0044, 0.0023, -0.0008, 0.0010, 0.0006]
vol_pos = [-0.0277, -0.0061, 0.0151, 0.0219, 0.0294]
```

Intuitively, higher MACD means higher short-term trend over long-term trend. This should imply lower prob of -1 class and higher prob of +1 class in short run. Then, in the long run, this kind of prediction would disappear. According to the diagram, for short-term window of forecast, higher MACD implies lower prob of -1 and higher prob of +1 in 3-day and 5-day. For longer window, this pattern will disappear (the marginal effect fluctuated around zero).

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, MACD_neg, WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('Average Marginal Effect')
plt.title('MACD: negative class')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, MACD_zero, WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('Average Marginal Effect')
plt.title('MACD: sideway class')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, MACD_pos, WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('Average Marginal Effect')
plt.title('MACD: positive class')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```

Momentum is defined as historical percentage rate of return. If momentum has increased, it is possible that there is still momentum to continue, so the prob of +1 should increase and prob of -1 should decrease (momentum effect). However, it is possible that the momentum will disappear soon because of reversal. Experts disagree on this. From the diagram, I see the momentum effect during 10-day, 14-day and 20-day window of forecast. In addition, the average marginal effect gradually diminishes from 10 to 20 day.

```python
WIDTH = 0.35
IDX = np.arange(5)

fig, ax = plt.subplots()
p1 = ax.bar(IDX, MOM_neg, WIDTH)
p2 = ax.axhline(y=0)

plt.ylabel('Average Marginal Effect')
plt.title('Momentum: negative class')
plt.xticks(IDX, ('3 days', '5 days', '10 days', '14 days', '20 days'))
```
